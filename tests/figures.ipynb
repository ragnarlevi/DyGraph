```python
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G = nx.Graph()
G.add_node(1)
G.add_node(2)
G.add_node(3)
G.add_node(4)
G.add_node(5)
G.add_edge(1,2)
G.add_edge(1,3)
G.add_edge(1,5)
G.add_edge(2,4)
G.add_edge(4,5)
G.add_edge(1,4)


edgelist = [(edge[0], edge[1]) for edge in G.edges(data = 'sign')]

pos = nx.spring_layout( G, weight=None)
```

```python
fig, ax = plt.subplots(1,1, figsize = (10,5))


nx.draw_networkx_nodes(G, pos, node_color= 'lightblue',ax = ax)
nx.draw_networkx_labels(G,pos,{i+1:i+1 for i in range(G.number_of_nodes())},ax = ax)
nx.draw_networkx_edges(
    G,
    pos,
    edgelist=edgelist,
    alpha=0.5,
    edge_color="tab:grey",
    ax = ax
)


x,y=pos[1]
ax.text(x+0.05,y+0.1,s=r'$f_1(x_t^{1})$',horizontalalignment='center',fontdict={'fontsize':15})
x,y=pos[2]
ax.text(x+0.05,y+0.1,s=r'$f_2(x_t^{2})$',horizontalalignment='center',fontdict={'fontsize':15})
x,y=pos[3]
ax.text(x+0.1,y,s=r'$f_3(x_t^{3})$',horizontalalignment='center',fontdict={'fontsize':15})
x,y=pos[4]
ax.text(x+0.05,y+0.1,s=r'$f_4(x_t^{4})$',horizontalalignment='center',fontdict={'fontsize':15})
x,y=pos[5]
ax.text(x,y+0.1,s=r'$f_5(x_t^{5})$',horizontalalignment='center',fontdict={'fontsize':15})
```
